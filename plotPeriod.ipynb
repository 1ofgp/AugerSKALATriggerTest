```python
import matplotlib.pyplot as plt
import numpy as np
import io
print("import done")
```

```python
filename = "/home/novikov/Downloads/functionGenerator_labTest_1000Hz.txt"
data = np.loadtxt(filename)
```

```python
# 3. Extract columns
# Column 0 is the event number, Column 1 is the timestamp
timestamps = data[:, 1]

# 4. Calculate Periods (Delta t)
# np.diff calculates the difference between consecutive elements: t[i] - t[i-1]
periods = np.diff(timestamps)

# Print some basic stats to console
print(f"Number of periods calculated: {len(periods)}")
print(f"Mean period: {np.mean(periods):.2f}")
print(f"Min period: {np.min(periods)}")
print(f"Max period: {np.max(periods)}")

# 5. Plot the Distribution
plt.figure(figsize=(6, 6))

# Create a histogram
# bins='auto' lets numpy decide the best bucket width
plt.hist(periods, bins='auto')

plt.title(filename)
plt.xlabel('period [ns]')
# plt.ylabel('Frequency')
# plt.grid(axis='y', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
